##Z score normalization
Creating a normal standard distribution from an existing set of data.


```python
import numpy as np
```

```python
data = np.array([10,20,30,60,80,60])
mean = np.mean(data)
std = np.std(data)

print(mean)
print(std)
```

Computing the Z score for each value of data

```python
dataNew = np.empty(data.shape[0])
for i in range(data.shape[0]):
  dataNew[i] = (data[i] - mean)/std

print(dataNew)
print(np.mean(dataNew))
print(np.std(dataNew))
```
